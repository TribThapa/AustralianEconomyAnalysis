# Employement data clean up

```python
#Import libraries and dependencies

import pandas as pd
import os
import csv
from pathlib import Path
```

```python
# Read in the CSV as Dataframe
Employment_df = pd.read_csv("../Data/RawData/Employment_JuneOnly.csv")

Employment_df.head()
```

```python
#Check the dataframe type
Employment_df.dtypes
```

```python
#Drop the unnamed/first column
#Employment_df.loc[:, ~Employment_df .columns.str.contains('^Unnamed')]
#Employment_df    
Employment_df.drop(Employment_df .columns[Employment_df .columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
Employment_df.head()
```

```python
#drop July and only keep June 
employment_june_df = Employment_df[Employment_df.Time.str.match('Jun')]
employment_june_df.head()
```

```python
#Only take 1995 to 2020 
Emp95_20=employment_june_df.iloc[17:]
Emp95_20.head()
```

```python
#Change Row name from Time to Year

rows = ['1995', '1996', '1997', '1998',
       '1999', '2000', '2001', '2002', '2003', '2004', '2005',
       '2006', '2007', '2008', '2009', '2010', '2011', '2012',
       '2013', '2014', '2015', '2016', '2017', '2018', '2019',
       '2020']

Emp95_20["Time"] = rows
Emp95_20.head()
```

```python
#Change column name from Time to Year
f_employment=Emp95_20.rename(columns = {'Time':'Year'})
f_employment.head()
```

```python
f_employment.columns
```

```python
final_employment=f_employment.set_index("Year")
final_employment.head()
```
