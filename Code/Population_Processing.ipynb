# Population data clean up

```python
#Import libraries and dependencies
import pandas as pd
```

```python
pop_df = pd.read_csv("../Data/RawData/Population.csv")
```

```python
# Get data for June only
pop_june_df = pop_df[pop_df.Time.str.match('Jun*')]

pop_june_df.to_csv("../Data/ProcessedData/Population_JuneOnly.csv")
```

```python
pop_june_df
```

```python
pop_june_df = pop_june_df.iloc[14:40] 
pop_june_df
```

```python
pop_june_df = pop_june_df.rename(columns={"Time" : "Year"})
pop_june_df
```

```python
rows = ["1995", "1996", "1997", "1998", "1999",
           "2000", "2001", "2002", "2003", "2004",
           "2005", "2006", "2007", "2008", "2009", "2010",
           "2011", "2012", "2013","2014", "2015", "2016", "2017",
           "2018", "2019", "2020"]
pop_june_df["Year"] = rows
```
