```python
#Import libraries and dependencies

import pandas as pd
import os
import csv
from pathlib import Path
```

```python
#Set the file path

btcus_csvpath = Path("../Data/RawData/BTC_USD_history.csv")

#Read in the CSV as Dataframe

BTC_USD_df = pd.read_csv(btcus_csvpath)

BTC_USD_df.head()
```

```python
#Filter to get June data only

BTC_USD_df = BTC_USD_df[BTC_USD_df.Date.str.match('Jun-*')]
```

```python
BTC_USD_df.head()
```

```python
#Drop unused columns
BTC_USD_df = BTC_USD_df.drop(BTC_USD_df.columns[2:], axis=1)
```

```python
BTC_USD_df.head()
```

```python
BTC_USD_df = BTC_USD_df.set_index("Date")
```

```python
# Write output as csv file
BTC_USD_df.to_csv("../Data/ProcessedData/BTC_USD_df.csv")
```
