```python
import pandas as pd
```

```python
# Load raw world data file
world_data = pd.read_csv("../Data/RawData/World_data.csv")
```

```python
# Check dataframe
world_data.head()
```

### Clean World data

```python
# Extract GDP data
au_vs_world_gdp = world_data[["year", "AU_gdp", "CA_gdp", "CN_gdp", "GB_gdp", "IN_gdp", "JP_gdp", "KR_gdp", "US_gdp"]]

# Remove '_gdp' string
au_vs_world_gdp.columns = au_vs_world_gdp.columns.str.rstrip('_gdp')

# Rename year column
au_vs_world_gdp.rename(columns={'year': "Year"}, inplace=True)

# Drop NAs
au_vs_world_gdp.dropna(inplace=True)

# Save to csv
au_vs_world_gdp.to_csv("../Data/ProcessedData/aus_vs_world_gdp_cleaned.csv")
```

```python
# Extract GDP growth data
au_vs_world_gdp_growth = world_data[["year", "AU_gdp_growth", "CA_gdp_growth", "CN_gdp_growth", "GB_gdp_growth", "IN_gdp_growth", "JP_gdp_growth", "KR_gdp_growth", "US_gdp_growth"]]

# Remove '_gdp_growth' string
au_vs_world_gdp_growth.columns = au_vs_world_gdp_growth.columns.str.strip('gdp_growth')

# Rename year column
au_vs_world_gdp_growth.rename(columns={'yea': "Year"}, inplace=True)

# Drop NAs
au_vs_world_gdp_growth.dropna(inplace=True)

# Save to csv
au_vs_world_gdp_growth.to_csv("../Data/ProcessedData/aus_vs_world_gdp_growth_cleaned.csv")
```

```python
# Extract world_goods_export data
au_vs_world_goods_export = world_data[["year", "AU_goods_export", "CA_goods_export", "CN_goods_export", "GB_goods_export", "IN_goods_export", "JP_goods_export", "KR_goods_export", "US_goods_export"]]

# Remove '_goods_export' string
au_vs_world_goods_export.columns = au_vs_world_goods_export.columns.str.rstrip('_goods_export')

# Rename year column
au_vs_world_goods_export.rename(columns={'yea': "Year"}, inplace=True)

# Drop NAs
au_vs_world_goods_export.dropna(inplace=True)

# Save to csv
au_vs_world_goods_export.to_csv("../Data/ProcessedData/aus_vs_world_goods_export_cleaned.csv")
```
