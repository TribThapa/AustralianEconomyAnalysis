```python
#Import libraries and dependencies

import pandas as pd
import os
import csv
from pathlib import Path
```

```python
#Set the file path

sp500_csvpath = Path("../Data/RawData/SP500_history.csv")

#Read in the CSV as Dataframe

SP500_history_df = pd.read_csv(sp500_csvpath)

SP500_history_df.head()
```

```python
#Filter to get June data only

SP500_history_df = SP500_history_df[SP500_history_df.Date.str.match('Jun-*')]
```

```python
SP500_history_df.head()
```

```python
#Drop unused columns
SP500_history_df = SP500_history_df.drop(SP500_history_df.columns[2:], axis=1)
```

```python
SP500_history_df.head()
```

```python
SP500_history_df = SP500_history_df.set_index("Date")
```

```python
# Write output as csv file
SP500_history_df.to_csv("../Data/ProcessedData/SP500_history_updated.csv")
```
