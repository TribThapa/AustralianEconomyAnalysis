```python
#Import libraries and dependencies
import pandas as pd
import os
import csv
from pathlib import Path
```

```python
# #Read in the CSV as Dataframe
GDP_countries_df = pd.read_csv("../Data/ProcessedData/GDP_selected_countries.csv")

GDP_countries_df.head()
```

```python
#Drop Unnamed column
GDP_countries_df = GDP_countries_df.drop(GDP_countries_df.columns[0], axis=1)
```

```python
GDP_countries_df.head()
```

```python
#Drop unused columns
GDP_countries_df = GDP_countries_df.drop(GDP_countries_df.columns[1:39], axis=1)
```

```python
GDP_countries_df.head()
```

```python
#Edit columns name to get the right format
GDP_countries_df.columns

columns = ['Year', '1995', '1996', '1997', '1998',
       '1999', '2000', '2001', '2002', '2003', '2004', '2005',
       '2006', '2007', '2008', '2009', '2010', '2011', '2012',
       '2013', '2014', '2015', '2016', '2017', '2018', '2019',
       '2020']

GDP_countries_df.columns = columns
```

```python
GDP_countries_df.head()
```

```python
#Transpose data
GDP_countries_df = GDP_countries_df.T
```

```python
GDP_countries_df.head()
```

```python
# Format for countries as headers
GDP_countries_df.columns = GDP_countries_df.iloc[0]
GDP_countries_df = GDP_countries_df[1:]
```

```python
GDP_countries_df
```

```python
#Remove 2020 N/A values
GDP_countries_df.drop(GDP_countries_df.tail(1).index,inplace=True)
```
